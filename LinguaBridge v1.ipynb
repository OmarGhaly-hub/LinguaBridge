```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def analyze_sentence(sentence):
    doc = nlp(sentence)
    observations = []
    
    # Check for copula (am/is/are)
    for token in doc:
        if token.lemma_ == "be" and token.dep_ == "ROOT":
            observations.append(
                "- Copula dropping: Your sentence contains 'am/is/are' as a main verb. In Arabic present tense, this verb is dropped entirely."
            )
            break
    
    # Check for definite article "the"
    for token in doc:
        if token.text.lower() == "the":
            observations.append(
                "- Definite article: 'the' in English becomes 'ال' in Arabic, attached directly to the start of the word it modifies."
            )
            break
    
    # Check for indefinite article "a/an"
    for token in doc:
        if token.text.lower() in ["a", "an"]:
            observations.append(
                "- Indefinite article: 'a/an' in English has no equivalent in Arabic. Arabic nouns are indefinite by default without any article."
            )
            break
    
    # Check for adjective-noun pairs
    for token in doc:
        if token.pos_ == "ADJ" and token.head.pos_ == "NOUN":
            observations.append(
                "- Adjective-noun order: English places adjectives before nouns. In Arabic, the noun comes first, followed by the adjective."
            )
            break
    
    # Check for plural nouns
    for token in doc:
        if token.tag_ == "NNS" or token.tag_ == "NNPS":
            observations.append(
                "- Plural distinction: English treats any quantity above one as plural. Arabic has three forms, singular, dual (Muthanna, for exactly 2), and plural (for 3 or more), each with its own grammatical rules."
            )
            break
    
    # Check for subject-verb order
    for token in doc:
        if token.dep_ == "nsubj":
            observations.append(
                "- Subject-verb order: English follows Subject-Verb-Object order. Arabic commonly uses Verb-Subject-Object order, placing the verb before the subject."
            )
            break
    # Check for grammatical gender (gendered pronouns)
    gendered_pronouns = ["he", "she", "his", "her", "him", "hers"]
    for token in doc:
        if token.text.lower() in gendered_pronouns:
           observations.append(
               "- Grammatical gender: English marks gender mainly in pronouns. In Arabic, gender applies to all nouns, adjectives, and verbs; every element in a sentence must agree grammatically in gender."
           )
           break

   # Check for pronoun as object of a verb
    for token in doc:
       if token.pos_ == "PRON" and token.dep_ in ["dobj", "dative"]:
           observations.append(
               "- Object pronoun attachment: In English, object pronouns like 'him/her/them' are separate words. In Arabic, they are attached as suffixes directly to the end of the verb."
           )
           break
    
    return observations
```

```python
def run_analyzer():
    print("=" * 60)
    print("English-Arabic Linguistic Contrast Analyzer")
    print("=" * 60)
    
    sentence = input("\nEnter an English sentence: ")
    
    print(f"\nAnalyzing: '{sentence}'")
    print("-" * 60)
    
    observations = analyze_sentence(sentence)
    
    if observations:
        print(f"\n{len(observations)} linguistic contrast(s) detected:\n")
        for obs in observations:
            print(obs)
            print()
    else:
        print("\nNo linguistic contrasts detected for this sentence.")
        print("Try a sentence with 'the', 'a/an', 'is/are', or adjectives.")
    
    print("=" * 60)

# Run it
run_analyzer()
```
